```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

# Exercício 1

Dado um dataset X = [peso] (em gramas) e Y = tamanho do braço (milimetros) para pinguins, faça uma previsão de qual seria a o tamanho do braço de um pinguim de uma tonelada.

```python
penguins = sns.load_dataset('penguins').dropna()
X = np.array(penguins['body_mass_g']).reshape(-1, 1)
Y = np.array(penguins['flipper_length_mm'])

plt.scatter(X, Y)
```

```python
# Codar
```

## Gabarito

```python
from sklearn import linear_model
reg = linear_model.LinearRegression()

reg.fit(X, Y)
tamanho = reg.predict([[1_000_000]])[0]
print(f'Tamanho do braço de um pinguim de uma tonelada: {tamanho/1000:.2f} metros')
```

```python
geyser = sns.load_dataset('geyser').dropna()
X = geyser[['duration', 'waiting']].to_numpy()
Y = geyser['kind']

plt.scatter(X[:, 0], X[:, 1], c=['red' if k=='long' else 'blue' for k in Y])
```

```python
# Codar
```

## Gabarito

```python
from sklearn import svm
clf = svm.SVC(kernel='linear')
clf.fit(X, Y)
clf.score(X, Y)
```

```python
X1, X2 = np.meshgrid(np.linspace(1.5, 5, 100), np.linspace(40, 100, 100))
Z = clf.predict(np.c_[X1.ravel(), X2.ravel()])
plt.scatter(X1, X2, c=['coral' if k=='long' else 'lightskyblue' for k in Z])
plt.scatter(X[:,0], X[:,1], c=['red' if k=='long' else 'blue' for k in Y])
```
